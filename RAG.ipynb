```python
import fitz  # PyMuPDF
import pdfplumber
import pandas as pd

# step 1 -- to Extract the text from the PDF i.e. from the provided doc
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page_num in range(doc.page_count):
        page = doc.load_page(page_num)
        text += page.get_text("text")
    return text

# Extracting tables using pdfplumber
def extract_tables_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        tables = []
        for page in pdf.pages:
            tables.extend(page.extract_tables())
    return tables


pdf_path = "bajaj_finserv_factsheet_Oct.pdf"
text_data = extract_text_from_pdf(pdf_path)
tables_data = extract_tables_from_pdf(pdf_path)
```

```python
from PyPDF2 import PdfReader
from langchain.document_loaders import TextLoader

# Extract text from the PDF
pdf_path = "/content/bajaj_finserv_factsheet_Oct.pdf"
reader = PdfReader(pdf_path)
text_data = ""

for page in reader.pages:
    text_data += page.extract_text()

# Now you can load the extracted text into TextLoader
text_loader = TextLoader(text_data)
docs = text_loader.load()
```

```python
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import FAISS
from langchain.text_splitter import CharacterTextSplitter
from langchain.document_loaders import TextLoader

# Load the text data extracted from the PDF
text_loader = TextLoader(text_data)  # Assuming text_data is the extracted text from the PDF
docs = text_loader.load()

# Split the text into chunks (this helps manage large documents)
text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=0)
chunks = text_splitter.split_documents(docs)

# Initialize OpenAI embeddings
embeddings = OpenAIEmbeddings()

# Create FAISS index
vector_store = FAISS.from_documents(chunks, embeddings)

# Example of retrieving a document (chunk) by user query
query = "What is the 3-year return of Bajaj Flexi Cap Fund?"
results = vector_store.similarity_search(query, k=3)
print(results)


pdf_path = "bajaj_finserv_factsheet_Oct.pdf"
text_data = extract_text_from_pdf(pdf_path)
tables_data = extract_tables_from_pdf(pdf_path)
```

```python
# Calculate Compound Annual Growth Rate (CAGR)
def calculate_cagr(initial_value, final_value, years):
    return ((final_value / initial_value) ** (1 / years)) - 1

# Example calculation for 3-year return
initial_value = 100  # This is just an example; replace with actual data from the factsheet
final_value = 130
years = 3
cagr = calculate_cagr(initial_value, final_value, years)
print(f"CAGR over 3 years: {cagr * 100:.2f}%")
```

```python
import streamlit as st
from langchain.chat_models import ChatOpenAI

# Function to process user queries
def answer_query(query):
    results = vector_store.similarity_search(query, k=3)
    answer = "Based on the factsheet, here is the answer:\n"
    for result in results:
        answer += f"Source: {result.metadata['source']}\n"
        answer += f"Answer: {result.page_content}\n"
    return answer

# Streamlit UI
st.title("Bajaj AMC Fund Factsheet Chatbot")
st.write("Ask me anything about the Bajaj AMC Fund Factsheet for Oct 2025.")

# User input for query
user_query = st.text_input("Enter your query:")

if user_query:
    response = answer_query(user_query)
    st.write(response)
```
